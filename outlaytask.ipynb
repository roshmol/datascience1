```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
data
```

```python
def function(data,column_name):
    q1=data[column_name].quantile(0.25)
    q2=data[column_name].quantile(0.50)
    q3=data[column_name].quantile(0.75)
    IQR=q3-q1
    lower_limit=q1-1.5*IQR
    upper_limit=q3+1.5*IQR
    return lower_limit,upper_limit
```

```python
data.columns
```

```python
data1=['Age','Annual Income (k$)','Spending Score (1-100)']
outliers={}
for i in data1:
    lower_limit,upper_limit=function(data,i)
    outliers[i]=[lower_limit,upper_limit]
```

```python
outliers
```

```python
fillrow=[]
for i in data1:
    fillrow.append(data[(data[i]<outliers[i][0])|(data[i]>outliers[i][1])].index.values)
```

```python
for k in range(len(fillrow)):
    data.loc[fillrow[k],data1[k]]=data[data1[k]].mean()
```
